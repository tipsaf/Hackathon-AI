# Génération de prompt / réponses

Certains des dizianes de prompts générés ont ensuite été corrigés et ameliorés à la main par nos soin pour améliorer le set de Question / réponses qu'on donne pour l'entrainement.

```python
df = pd.read_csv("enhanced_unique_plants_dataset.csv")

# Calculer les prix minimum et maximum dynamiquement
min_price = df["Price (EUR)"].min()
max_price = df["Price (EUR)"].max()

# Vérification des données valides
def is_valid_row(row):
    required_columns = ["Plant Name", "Water Needs", "Light Needs", "Soil Type", "Description", "Price (EUR)", "Rating (out of 5)"]
    return all(pd.notnull(row[col]) for col in required_columns)

# Inférer l'attribut d'une plante
def infer_attribute(description):
    if "tropical" in description.lower():
        return "tropical"
    if "colorful" in description.lower():
        return "colorful"
    if "indoor" in description.lower():
        return "indoor"
    return "decorative"

# Appliquer des règles intelligentes à la requête
def apply_query_rules(prompt, row):
    # Initialisation des règles
    rules = {
        "bright spaces": row["Light Needs"] == "High",
        "decorate": "colorful" in row["Description"].lower(),
        "small plant": row["Price (EUR)"] <= 10,
        "cheap plant": row["Price (EUR)"] < 10,
        "expensive plant": row["Price (EUR)"] > 20,
        "good gift": "beautiful" in row["Description"].lower() or "elegant" in row["Description"].lower(),
        "easy to maintain": row["Water Needs"] == "Low"
    }

    # Appliquer les règles au prompt
    for key, condition in rules.items():
        if key in prompt.lower() and not condition:
            return False
    return True

# Templates pour les réponses
response_templates = [
    "{name} is an excellent choice for {context}. It thrives in {soil} soil and requires {light} light and {water} water. It is priced at {price} euros with a rating of {rating}/5.",
    "{name} is perfect for {context}. It needs {light} light, {water} water, and grows best in {soil} soil. It costs {price} euros and is rated {rating}/5.",
    "{name} stands out as a great option for {context}. With {light} light, {water} water, and {soil} soil requirements, it is priced at {price} euros and rated {rating}/5."
]

# Liste des questions pour générer des prompts
questions = [
    "I need a plant that needs {water} water needs and {light} light needs.",
    "Suggest a {attribute} plant with {water} water needs.",
    "Can you recommend a {attribute} plant that thrives in {soil} soil?",
    "What is a good plant for {environment}?",
    "I'm looking for a {size} plant for my {place} that is easy to maintain."
]

# Contextes pour remplir les variables dans les questions
environments = ["low light conditions", "air purification", "bright spaces"]
sizes = ["small", "medium-sized", "large"]
places = ["office desk", "living room", "bedroom"]

# Générer des prompts et réponses
prompts_responses = []
for _, row in df.iterrows():
    if not is_valid_row(row):
        continue  # Ignorer les lignes invalides

    # Générer un attribut basé sur la description
    attribute = infer_attribute(row["Description"])

    # Générer un prompt aléatoire
    prompt_template = random.choice(questions)
    prompt = prompt_template.format(
        attribute=attribute,
        water=row["Water Needs"].lower(),
        light=row["Light Needs"].lower(),
        soil=row["Soil Type"].lower(),
        environment=random.choice(environments),
        size=random.choice(sizes),
        place=random.choice(places),
    )

    # Appliquer les règles intelligentes
    if not apply_query_rules(prompt, row):
        continue  # Ignorer cette plante si elle ne correspond pas aux règles

    # Générer une réponse aléatoire
    response_template = random.choice(response_templates)
    response = response_template.format(
        name=row["Plant Name"],
        context=random.choice(environments),
        soil=row["Soil Type"],
        light=row["Light Needs"],
        water=row["Water Needs"],
        price=row["Price (EUR)"],
        rating=row["Rating (out of 5)"],
    )

    # Ajouter le prompt et la réponse
    prompts_responses.append({"prompt": prompt, "response": response})

# Afficher quelques exemples
for example in random.sample(prompts_responses, 10):
    print(f"Prompt: {example['prompt']}\nResponse: {example['response']}\n")
```
